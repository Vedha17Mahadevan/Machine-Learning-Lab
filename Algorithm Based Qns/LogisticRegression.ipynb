```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, precision_score
```

```python
df = pd.read_csv('student_data.csv')
print(df)
```

```python
 #define targets and features
X = df[['study_hours']] 
y = df['pass_exam']

#splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#LOGISTIC REGRESSION

model = LogisticRegression()
model.fit(X_train, y_train)

#predictions
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

#training data set
acc_train = accuracy_score(y_train, y_train_pred)
cm_train = confusion_matrix(y_train, y_train_pred)

print("Training Data Set")
print("Accuracy: ", acc_train)
print("Confusion Matrix: \n", cm_train)
print("Classification Report: \n", classification_report(y_train, y_train_pred))

#visualization training
plt.figure(figsize=(5,8))
sns.heatmap(cm_train, annot = True, fmt = 'd', color='blue')
plt.xlabel('Study Hours')
plt.ylabel('Pass Exam')
plt.title('Training Dataset Heatmap for Confusion Matrix')
plt.legend()
plt.show()

#test data set
acc_test = accuracy_score(y_test, y_test_pred)
cm_test = confusion_matrix(y_test, y_test_pred)

print("Test Data Set")
print("Accuracy: ", acc_test)
print("Confusion Matrix: \n", cm_test)
print("Classification Report: \n", classification_report(y_test, y_test_pred))

#visualization training
plt.figure(figsize=(5,8))
sns.heatmap(cm_test, annot = True, fmt = 'd', color='green')
plt.xlabel('Study Hours')
plt.ylabel('Pass Exam')
plt.title('Testing Dataset Heatmap for Confusion Matrix')
plt.legend()
plt.show()
```
