```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = fetch_california_housing()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

print(df.head(8))

r,c = df.shape
print("Rows: ",r)
print("Columns: ",c)

X=data.data
y=data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#knn

import numpy as np

def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2)**2))

def knn_regressor(X_train, y_train, X_test, k=5):
    preds = []

    for test_point in X_test:
        distances = []

        for i in range(len(X_train)):
            d = euclidean_distance(test_point, X_train[i])
            distances.append((d, y_train[i]))

        distances.sort(key=lambda x: x[0])
        neighbors = distances[:k]

        values = [value for _, value in neighbors]
        preds.append(np.mean(values))

    return np.array(preds)

y_pred_knn = knn_regressor(X_train, y_train, X_test, k=5)
print("MSE knn:", mean_squared_error(y_test, y_pred_knn))
print("R2 knn: ", r2_score(y_test, y_pred_knn))

# single linear regr
def fit_linear(X,y):
  x = X[:,7]
  x_mean = np.mean(x)
  y_mean = np.mean(y)

  num = np.sum((x-x_mean)*(y-y_mean))
  den = np.sum((x-x_mean)**2)

  m = num/den
  c = y_mean-m*x_mean
  return m,c
def predict_linear(X,m,c):
  x=X[:,7]
  return m*x+c
m, c = fit_linear(X_train, y_train)
y_pred_linear = predict_linear(X_test, m, c)

#multivariate
def fit_multivariate_regression(X, y):
    X_b = np.c_[np.ones((X.shape[0], 1)), X]
    beta = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y
    return beta

def predict_multivariate(X, beta):
    X_b = np.c_[np.ones((X.shape[0], 1)), X]
    return X_b @ beta

beta_multi = fit_multivariate_regression(X_train, y_train)
y_pred_multi = predict_multivariate(X_test, beta_multi)

#polynomial
def fit_polynomail_regression(X,y):
  x=X[:,7]
  X_poly = np.c_[np.ones(len(x)), x, x**2]
  beta = np.linalg.inc(X_poly.T@X_poly)@X_poly.T@y
  return beta

def predict_polynomail(X,beta):
  x=X[:,7]
  X_poly=np.c_[np.ones(len(x)),x,x**2]
  return X_poly@beta

  beta_poly = fit_polynomail_regression(X_train, y_train)
  y_pred_poly = predict_polynomail(X_test, beta_poly)

#cart
from sklearn.tree import DecisionTreeRegressor

cart = DecisionTreeRegressor(random_state=42)
cart.fit(X_train, y_train)
y_pred_cart = cart.predict(X_test)
print("CART -> MSE:", mean_squared_error(y_test, y_pred_cart), "R2:", r2_score(y_test, y_pred_cart))
```
