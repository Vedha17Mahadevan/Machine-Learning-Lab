```python
# 1.1: To implement a menu driven program using Numpy - Exercise 1
while(1):
    
    print("\n\nWelcome to Menu Driven Program for Familiarization with Numpy- Exercise I")
    print("\nMenu: ")
    print("1. Create a 1D array of numbers from 0-9 and display the first 3 elements, last 3 elements and elements from index 2 to 5.")
    print("2. Create a 3x3 matrix with values 0-8 and display the element at index [1,2] and elements in first 2 rows and columns.")
    print("3. Create a 3x3 identity matrix and flatten the 2D array back to 1D array.")
    print("4. Create a 1D array of integers from 1 top 10. Replace all odd nuumbers with -1.")
    print("5. Given an array of random numbers, compute mean, median, standard deviation and variance.")
    print("6. Exit")

    choice = int(input("Enter your choice (1/2/3/4/5/6): "))
    
    if(choice==1):
        arr_1D = np.arange(10),
        print("Original array: ", arr_1D),
        print("First 3 elements of the array: ",arr_1D[:3]),
        print("Last 3 elements of the array: ", arr_1D[-3:]),
        print("Elements between 2 to 6: ", arr_1D[2:6])
    elif(choice==2):
        mat_3x3 = np.arange(9).reshape(3,3)
        print("Matrix: ","\n", mat_3x3)
        print("Element at 2nd row 3rd column: ", mat_3x3[1,2])
        print("Element at first 2 rows and first 2 columns: ","\n", mat_3x3[:2, :2])
    elif(choice==3):
        id_matr = np.eye(3)
        id_flat = id_matr.flatten()
        print("Identity Matrix: ", "\n", id_matr)
        print("Flattened Identity Matrix: ", id_flat)
    elif(choice==4):
        int_arr = np.arange(1,11)
        print("Array: ", int_arr)
        int_arr[int_arr%2 != 0] = -1
        print("Replacing all odd numbers with -1: ", int_arr)
    elif(choice==5):
        rand_arr = np.random.rand(10)
        print("Mean: ", np.mean(rand_arr))
        print("Median: ", np.median(rand_arr))
        print("Standard Deviation: ", np.std(rand_arr))
        print("Variance: ", np.var(rand_arr))
    elif(choice==6):
        print("Exiting....")
        break
    else:
        print("Invalid choice... Try Again")
```

```python
# 1.2: To implement a menu driven program using Numpy - Exercise 2
while(1):
    
    print("\n\nWelcome to Menu Driven Program for Familiarization with Numpy- Exercise I")
    print("\nMenu: ")
    print("1. Create 2x2 matrix and perform multiplication.")
    print("2. Compute determinant and inverse of a 2x2 matrix.")
    print("3. Multiply 2D array by a 1D array.")
    print("4. Find Euclidean distance between 2 points.")
    print("5. Exit")

    choice = int(input("Enter your choice (1/2/3/4/5): "))
    
    if(choice==1):
        matr1 = np.arange(4).reshape(2,2)
        matr2 = np.arange(5,9).reshape(2,2)
        print("Original matrix 1: ", "\n", matr1)
        print("Original matrix 2: ", "\n", matr2)
        prod = np.dot(matr1,matr2)
        print("Product: ", "\n", prod)
    elif(choice==2):
        matrix = np.arange(4).reshape(2,2)
        print("Original matrix: ", "\n", matrix)
        print("Determinant: ", np.linalg.det(matrix))
        print("Inverse: ", "\n", np.linalg.inv(matrix))
    elif(choice==3):
        matrix_2D = np.array([[1,2],[3,4]])
        matrix_1D = np.array([1,2])
        print(" 2D Matrix: ", "\n", matrix_2D)
        print("1D Matrix: ", "\n", matrix_1D)
        print("Product: ", "\n", matrix_2D*matrix_1D[:,np.newaxis])
    elif(choice==4):
        p1 = 2
        p2 = 4

        dist = np.linalg.norm(p1-p2)
        print(dist)
    elif(choice==5):
        print("Exiting....")
        break
    else:
        print("Invalid choice... Try Again")
```
