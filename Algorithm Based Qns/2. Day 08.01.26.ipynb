```python
#1.4

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#1.4
df = pd.read_csv('Jobsearchdata.csv')
print(df)
```

```python
#1.4 a
print("Number of ROWS and COLUMNS: ", df.shape)
```

```python
#1.4 b

print("Candidate details for Each SkillSet: ", df.groupby('Skill set').apply(lambda x: x))
```

```python
#1.4 c

print("Job seekers whose salary is greater than 25,000: ", df[df['Salary']>25000])
```

```python
#1.4 d

print("Displaying Name and Years of Experience in Sorted Order: ", df[['Candidate Name','Years of Experience']].sort_values(by='Candidate Name'))
```

```python
#1.4 e

location_counts = df['Preferred Location'].value_counts()

plt.figure(figsize=(10, 6))
location_counts.plot(kind='bar')
plt.xlabel('Preferred Location')
plt.ylabel('Number of People')
plt.title('Distribution of People on Preferred Locations')

plt.xticks(rotation=45)

plt.show()
```

```python
#1.5
df = pd.read_csv('items.csv')
print(df)
```

```python
#1.5 d

print(df['Category'].value_counts())
```

```python
#1.6
df = pd.read_csv("Iris_Data.csv")
print(df)
```

```python
#1.6 a

print("The number of Data Points: ", df.shape)
print("\nThe Column names: \n", df.columns)
print("\nThe data type of each Column: \n", df.dtypes)
```

```python
#1.6 b

print("Removing Prefix - IRIS from species name: \n", df['species'].str.replace('Iris-',""))
```

```python
#1.6 c

print("\nCount of no. of species: \n",df['species'].value_counts())

print("\n----- SEPAL LENGTH-----")

print("\nCalculate MEAN of Sepal Length: \n",df['sepal length (cm)'].mean())

print("\nCalculate MEDIAN of Sepal Length: \n",df['sepal length (cm)'].median())

print("\nCalculate Quantiles of Sepal Length: \n",df['sepal length (cm)'].quantile([0.25, 0.5, 0.75]))

print("\nCalculate RANGE of Sepal Length: \n",df['sepal length (cm)'].max()-df['sepal length (cm)'].min())


print("\n----- SEPAL WIDTH-----")

print("\nCalculate MEAN of Sepal Width: \n",df['sepal width (cm)'].mean())

print("\nCalculate MEDIAN of Sepal Width: \n",df['sepal width (cm)'].median())

print("\nCalculate Quantiles of Sepal Width: \n",df['sepal width (cm)'].quantile([0.25, 0.5, 0.75]))

print("\nCalculate RANGE of Sepal Width: \n",df['sepal width (cm)'].max()-df['sepal width (cm)'].min())


print("\n----- PETAL LENGTH-----")

print("\nCalculate MEAN of Petal Length: \n",df['petal length (cm)'].mean())

print("\nCalculate MEDIAN of Petal Length: \n",df['petal length (cm)'].median())

print("\nCalculate Quantiles of Petal Length: \n",df['petal length (cm)'].quantile([0.25, 0.5, 0.75]))

print("\nCalculate RANGE of Petal Length: \n",df['petal length (cm)'].max()-df['petal length (cm)'].min())


print("\n----- PETAL WIDTH-----")

print("\nCalculate MEAN of Petal Width: \n",df['petal width (cm)'].mean())

print("\nCalculate MEDIAN of Petal Width: \n",df['petal width (cm)'].median())

print("\nCalculate Quantiles of Petal Width: \n",df['petal width (cm)'].quantile([0.25, 0.5, 0.75]))

print("\nCalculate RANGE of Petal Width: \n",df['petal width (cm)'].max()-df['petal width (cm)'].min())
```

```python
#1.6 d

print("\nCalculate MEAN : \n", df.groupby('species').mean())

print("\nCalculate MEDIAN: \n", df.groupby('species').median())
```

```python
#1.6 e

import seaborn as sns

sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='species', data=df)
plt.xlabel("Sepal Length (cm)")
plt.xlabel("Sepal Width (cm)")
plt.title("Scatter plot of Sepal Length vs Sepal Width")
```

```python
#1.6 f

import seaborn as sns

plt.hist(df['sepal length (cm)'], bins=20)
plt.xlabel("Sepal Length (cm)")
plt.xlabel("Frequency")
plt.title("Histogram od Sepal Length")
plt.show()
```
